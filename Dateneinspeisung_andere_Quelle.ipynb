## Einlesen der PDF-Texte

### EWMS

```python
#Initialisierung eines leeren Textes
EWMS=""
#Dokument mithilfe des PDF-Readers öffnen
with pdfplumber.open("Skript_EWMS_Sommer23.pdf") as pdf:
    for page in pdf.pages:
        #Text von den PDF-Seiten auslesen
        page_text =page.extract_text()
        #Text der jeweiligen Seiten an den leeren Text anhängen
        EWMS+=page_text
```

```python
#Ersetzen der Punkte aus dem Skript durch den Buchstaben e, da dies der häufigste Buchstabe im Alphabet ist
punkte=EWMS[5]
EWMS = EWMS.replace(punkte,'')
```

```python
EWMS = re.sub(r'[^a-zA-Z]', '', EWMS).lower()
```

```python
Paare=[EWMS[i:i+2] for i in range(len(EWMS)-1)]
AnzahlEWMS=Counter(Paare)
```

### Analysis1

```python
#Initialisierung eines leeren Textes
Ana=""
#Dokument mithilfe des PDF-Readers öffnen
with pdfplumber.open("Ana1Skript.pdf") as pdf:
    for page in pdf.pages:
        #Text von den PDF-Seiten auslesen
        page_text =page.extract_text()
        #Text der jeweiligen Seiten an den leeren Text anhängen
        Ana+=page_text
```

```python
Ana = Ana.replace(punkte,'')
Ana = re.sub(r'[^a-zA-Z]', '', Ana).lower()
```

```python
Paare=[Ana[i:i+2] for i in range(len(Ana)-1)]
AnzahlAna=Counter(Paare)
```

## Anlegen des dataFrames

```python
alphabet=list(map(chr, range(97, 123)))
```

```python
B2=[]
for i in range(26):
    for j in range (26):
        B2.append(alphabet[j])
B1=[]
for i in range(26):
    for j in range(26):
        B1.append(alphabet[i])
```

```python
Index=range(0,676)
```

```python
Buchstabe1=pd.Series(B1,index=Index)
Buchstabe2=pd.Series(B2,index=Index)
```

```python
Daten=pd.concat([Buchstabe1,Buchstabe2],axis=1)
Spalten=['Buchstabe1','Buchstabe2']
Daten.columns=Spalten
Daten['Buchstaben']=Daten['Buchstabe1']+Daten['Buchstabe2']
```

```python
Daten.head()
```

## absolute Häufigkeiten

```python
# EWMS
AnzE=[]
for i in range(len(Daten['Buchstaben'])):
    if AnzahlEWMS[Daten['Buchstaben'][i]]>0:
        AnzE.append(AnzahlEWMS[Daten['Buchstaben'][i]])
    else:
        AnzE.append(0)
```

```python
# Ana
AnzA=[]
for i in range(len(Daten['Buchstaben'])):
    if AnzahlAna[Daten['Buchstaben'][i]]>0:
        AnzA.append(AnzahlAna[Daten['Buchstaben'][i]])
    else:
        AnzA.append(0)
```

```python
Daten['absH_EWMS']=AnzE
Daten['absH_Ana']=AnzA
```

```python
Daten.head()
```

## relative Häufigkeiten

```python
#EWMS
sumE=0
relativeHäufE=[]

for i in range(len(Daten['Buchstaben'])):
    sumE+=Daten['absH_EWMS'][i]
for i in range(len(Daten['Buchstaben'])):
    relativeHäufE.append(Daten['absH_EWMS'][i]/sumE)
Daten['relH_EWMS']=relativeHäufE

#Ana
sumA=0
relativeHäufA=[]

for i in range(len(Daten['Buchstaben'])):
    sumA+=Daten['absH_Ana'][i]
for i in range(len(Daten['Buchstaben'])):
    relativeHäufA.append(Daten['absH_Ana'][i]/sumA)
Daten['relH_Ana']=relativeHäufA
```
