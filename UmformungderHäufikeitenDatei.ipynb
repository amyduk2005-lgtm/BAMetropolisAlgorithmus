```python
spalten_namen=['Buchstabe1','Buchstabe2', 'Quelle1_abs', 'Quelle2_abs', 'Quelle3_abs', 'Quelle4_abs', 'Quelle5_abs', 'Quelle1_rel', 'Quelle2_rel', 'Quelle3_rel', 'Quelle4_rel', 'Quelle5_rel']
data=pd.read_csv("BigramWhk_klein.txt",sep='\t',header=None, names=spalten_namen)
data.head()
```

```python
wkeit=data['Quelle1_rel']
```

```python
B1=data['Buchstabe1']
B2=data['Buchstabe2']
```

```python
neuerDataframe=pd.DataFrame(index=B2[0:26],columns=B2[0:26])
neuerDataframe.index.name=None
neuerDataframe.head()
```

```python
t=[]
for i in range (0,len(data),26):
    t.append(wkeit[i:i+26])
t = [list(x) for x in t]
```

```python
for i in range (26):
    neuerDataframe.loc[B2[0:26][i]]=t[i]
```
